### 분석 도전

미국 동북중부 437개 지역의 인구통계 정보를 담고 있는 midwest.csv를 사용해 데이터 분석 문제를 해결해 보세요.


```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('midwest.csv')
df.head()
```

#### Q1.

popadults는 해당 지역의 성인 인구, poptotal은 전체 인구를 나타냅니다.  
midwest 데이터에 '전체 인구 대비 미성년 인구 백분율' 변수를 추가하세요.


```python
new_df = df.assign(
    perchildren=(df['poptotal'] - df['popadults']) / df['poptotal'] * 100)
```

#### Q2.

미성년 인구 백분율이 가장 높은 상위 5개 country(지역)의 미성년 인구 백분율을 출력하세요.


```python
new_df.sort_values('perchildren', ascending=False).head(5)[['county', 'perchildren']]
```

#### Q3.

분류표의 기준에 따라 미성년 비율 등급 변수를 추가하고, 각 등급에 몇 개의 지역이 있는지 알아보세요.

---

분류 || 기준  
large || 40%  
middle || 30 ~ 40% 이상  
small || 30% 이상


```python
new_df = new_df.assign(childrengrade=np.where(new_df['perchildren'] >= 40, 'large',
                                              np.where(new_df['perchildren'] >= 30, 'middle', 'small')))
```

```python
new_df.groupby('childrengrade') \
    .agg(count=('childrengrade', 'count'))
```

```python
new_df['childrengrade'].value_counts()
```

#### Q4.
popasian은 해당 지역의 아시아인 인구를 나타냅니다.  
'전체 인구 대비 아시아인 인구 백분율' 변수를 추가하고, 하위 10개 지역의 state(주), country(지역), 아시아인 인구 백분율을 출력하세요.  

```python
df.assign(
    perasian=(df['popasian'] / df['poptotal']) * 100).sort_values('perasian').head(10)[['state', 'county', 'perasian']]
```
