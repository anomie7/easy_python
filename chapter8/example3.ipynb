```python
import pandas as pd
import numpy as np
import seaborn as sns
```

### Q1.
psavert(개인 저축률)가 시간에 따라 어떻게 변해 왔는지 알아보려고 합니다.  
연도별 개인 저축률의 변화를 나타낸 시계열 그래프를 만들어 보세요.


```python
economics = pd.read_csv('../economics.csv')
economics.head()
```

```python
economics['year'] = pd.to_datetime(economics['date']).dt.year
economics
```

```python
sns.lineplot(data = economics, x = 'year', y = 'psavert', ci=False)
```

### Q2.
2014년 월별 psavert의 변화를 나타낸 시계열 그래프를 만들어 보세요.

```python
df_new = economics.query('year == 2014')
df_new['month'] = pd.to_datetime(df_new['date']).dt.month
df_new
```

```python
sns.lineplot(data = df_new, x = 'month', y = 'psavert')
```
