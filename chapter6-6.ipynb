```python
import pandas as pd
import numpy as np
```

```python
exam = pd.read_csv('exam.csv')
```

```python
exam.agg(mean_math=('math', 'mean'))
```

```python
exam.groupby('nclass') \
    .agg(mean_math=('math', 'mean'))
```

```python
# 변수를 인덱스로 바꾸지 않기
exam.groupby('nclass', as_index=False) \
    .agg(mean_math=('math', 'mean'))
```

```python
exam.groupby('nclass') \
    .agg(mean_math=('math', 'mean'),
         sum_math=('math', 'sum'),
         median_math=('math', 'median'),
         n=('nclass', 'count'))
```

```python
# 모든 변수의 요약 통계량 한번에 구하기
exam.groupby('nclass').mean()
```

```python
mpg = pd.read_csv('mpg.csv')
mpg.info()
```

```python
mpg.groupby(['manufacturer', 'drv']) \
    .agg(mean_cty=('cty', 'mean'))
```

```python
# audi의 drv별 빈도
mpg.query('manufacturer == "audi"') \
    .groupby(['drv']) \
    .agg(n=('drv', 'count'))
```

```python
mpg.query('manufacturer == "audi"')['drv'] \
    .value_counts() \
    .to_frame('n') \
    .rename_axis('drv')
```
