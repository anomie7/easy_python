```python
import pandas as pd
import numpy as np
```

```python
# 결측치 만들기
df = pd.DataFrame({
    'sex': ['M', 'F', np.nan, 'M', 'F'],
    'score': [5, 4, 3, 4, np.nan]
})
df
```

```python
df['score'] + 1
```

```python
pd.isna(df)
```

```python
pd.isna(df).sum()
```

```python
# score 결측치 제거하기
df_nomiss = df.dropna(subset=['score'])
df_nomiss['score'] + 1
```

```python
df_nomiss = df.dropna(subset=['score', 'sex'])
df_nomiss
```

```python
df_nomiss2 = df.dropna()
df_nomiss2
```

```python
# pd.mean(), pd.sum()과 같은 수치 연산 함수는 결측치가 있으면 자동으로 제거하고 연산하는 기능이 있음
# groupby(), agg()를 이용해 집단별 요약 통계량을 구할 때도 결측치를 제거하고 연산합니다.
df['score'].mean(), df['score'].sum()
```

```python
df.groupby('sex').agg(mean_score=('score', 'mean'),
                      sum_score=('score', 'sum'))
```

```python
# 평균값으로 결측치 대체하기
exam = pd.read_csv('../exam.csv')
exam.loc[[2, 7, 14], ['math']] = np.nan
exam
```

```python
exam['math'].mean()
```

```python
# 결측치를 평균값으로 대체
exam['math'] = exam['math'].fillna(55)
exam
```
